1. count plots(used for categorial data)

count plot are used when you have categorial variables , and you want to visualize the distribution of these categories.

```python
sns.countplot(x='class',data=df)
```

```python
sns.countplot(y='class',data=df)
```

```python
sns.countplot(x='source_city',data=df)
```

```python
sns.countplot(y='source_city',data=df)
```

2. Piechart (used for categrial data)
A pie chart prepersents the parts of a whole entity.

```python
df['source_city'].value_counts().plot(kind='pie',autopct='%.2f')
```

Donut plots (used for categorial; data)

Donut plot are used to show the proprtion of each caegory in relation to the whole.

```python
plt.pie(df['source_city'].value_counts(),
        labels=df['source_city'].unique(),
        autopct='%.2f',wedgeprops=dict(width=0.8))
```

Histogram (used for numerical data) (continous variables)

Histograms are used when you have a continous variable,and you want to visualize its distribution

```python
plt.hist(df['price'])
```
